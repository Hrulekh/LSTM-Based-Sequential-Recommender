```python
import pandas as pd

ratings = pd.read_csv(
    "../data/ml-1m/ratings.dat",
    sep="::",
    engine="python",
    names=["user","movie","rating","time"]
)

movies = pd.read_csv(
    "../data/ml-1m/movies.dat",
    sep="::",
    engine="python",
    names=["movie","title","genre"],
    encoding="latin-1"
)

print(ratings.head())
print(movies.head())
```

```python
print("Number of users:", ratings.user.nunique())
print("Number of movies:", ratings.movie.nunique())
print("Number of ratings:", len(ratings))
```

```python
# Convert ratings into implicit feedback
ratings["label"] = (ratings["rating"] >= 4).astype(int)

ratings.head()
```

```python
ratings = ratings[ratings["label"] == 1]
print("Remaining interactions:", len(ratings))
```

```python
user_map = {u:i for i,u in enumerate(ratings.user.unique())}
movie_map = {m:i for i,m in enumerate(ratings.movie.unique())}

ratings["user_idx"] = ratings["user"].map(user_map)
ratings["movie_idx"] = ratings["movie"].map(movie_map)

n_users = len(user_map)
n_movies = len(movie_map)

print("Total users:", n_users)
print("Total movies:", n_movies)
```

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(ratings, test_size=0.2, random_state=42)

print("Train size:", len(train_df))
print("Test size:", len(test_df))
```
